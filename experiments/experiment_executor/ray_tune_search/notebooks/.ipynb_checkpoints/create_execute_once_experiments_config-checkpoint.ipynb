```python
def read_csv(dataset,type_simclr,n_transformations,percentage,dir):
    
    try:
        if(n_transformations>-1):
            folder=f'../{dir}{type_simclr}/{dataset}/'
            folder_percentage=f'{type_simclr}_{n_transformations}T_{dataset}_P{percentage}/data.csv'
        else:
            folder=f'../{dir}/{dataset}/'
            folder_percentage=f'{type_simclr}_{dataset}_P{percentage}/data.csv'
        
        data = pd.read_csv(f'{folder}{folder_percentage}')
        data.loc[data['score'] < 0,'score'] = 0
    except:
        print(f'{folder}    {folder_percentage}')
    
        data = pd.DataFrame({'Unnamed: 0': [], 'score': []})
    return data
    
```

```python
import yaml
import pandas as pd
import pandas as pd
datasets = ['kuhar', 'motionsense', 'uci']

models=['simclr','simclr_full','simclr_linear']

percentages=[25, 50, 75, 100, 200]
n_transformations_list=[1,2,3,4]
def get_max_by_each_transformation():
    dir="experiments/simclr_all/by_n_transformations/"
    
    df_maximas = pd.DataFrame() 
    for dataset in datasets:       
            for i, percentage in enumerate(percentages):
                for n_transformations in n_transformations_list:
                    max_length = 0
                    for j, model in enumerate(models):
                        try:
                            data = read_csv(dataset, model, n_transformations, percentage,dir)
                            max_dataset_score = data['score'].max()
                            indice_maximo = data['score'].idxmax()
                            fila_maxima = data.loc[indice_maximo].copy()  # Create a copy to avoid SettingWithCopyWarning
                            fila_maxima['percentage']=percentage
                            fila_maxima['n_transformations']=n_transformations
                            fila_maxima['dir']=str(dir)
                            fila_maxima['dataset']=dataset
                            fila_maxima['model']=model
                            fila_maxima['test']='sep_transf'

                            df_maximas = pd.concat([df_maximas, fila_maxima.to_frame().T], ignore_index=True)
    
                        except:
                            e=1
    return df_maximas
           

df_each_tranf=get_max_by_each_transformation()
    
    
                   
df_each_tranf                
```

```python
def get_max_all_transf():
    test_list=['tem_transf','best_transf','tem_transf']
    #test_list=['best_transf']
    df_maximas = pd.DataFrame()       
    for dataset in datasets:           
            for i, percentage in enumerate(percentages):
             for model in models:
                for test in test_list:
                    dir=f'experiments/simclr_all/all_transformations/{model}/{test}'
   
                    try:                
                        n_transformations=-1
                        data = read_csv(dataset, model, n_transformations, percentage,dir)
                        max_dataset_score = data['score'].max()
                        indice_maximo = data['score'].idxmax()
                        fila_maxima = data.loc[indice_maximo].copy()  # Create a copy to avoid SettingWithCopyWarning
                        fila_maxima['percentage']=percentage
                        fila_maxima['n_transformations']=n_transformations
                        fila_maxima['dir']=str(dir)
                        fila_maxima['dataset']=dataset
                        fila_maxima['model']=model  
                        fila_maxima['test']=test
                        
                        df_maximas = pd.concat([df_maximas, fila_maxima.to_frame().T], ignore_index=True)
                   
                    except:
                       e=1
    return df_maximas
df_all_tf=get_max_all_transf()
df_all_tf
```

```python
def get_df_fmax_score(groupby):
    idx_max = df_all_best.groupby(groupby)['score'].idxmax()
    df_fmax_score = df_all_best.loc[idx_max]
    return df_fmax_score
```

```python
def get_combined_data(df_fmax_score):
    combined_data = pd.DataFrame()
    for index, row in df_fmax_score.iterrows():
        dataset=row['dataset']
        model=row['model']
        n_transformations=row['n_transformations']
        percentage=row['percentage']
        dir=row['dir']
        #print(dir,dataset,n_transformations,percentage,model)
        data = read_csv(dataset, model, n_transformations, percentage,dir)
        data['dataset'] = dataset
        data['model'] = model
        data['n_transformations'] = n_transformations
        data['percentage'] = percentage
        combined_data = pd.concat([combined_data, data], ignore_index=True)
    combined_data.columns
    return combined_data
```

```python
import math
import numpy as np
import re
def create_folder(folder_name):
    import os
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
        
def new_base_config(exploration_file,config,new_file):
    with open(exploration_file, 'r') as archivo:
        datos_yaml = yaml.safe_load(archivo)


    for key, value in config.items():
            if(str(value)!='nan'):
                print(key,value)
                if isinstance(value, np.generic):
                    value = value.item()
                datos_yaml['reducer']['kwargs'][key] = value

    with open(new_file, 'w') as nf:
        yaml.dump(datos_yaml, nf, default_flow_style=False)
    
    return datos_yaml


def get_row_max_by_dataset(combined_data):
    idx_max = combined_data.groupby(['dataset'])['score'].idxmax()
    fmax_score = combined_data.loc[idx_max]    
    return fmax_score




def create_best_config_exploration(test):
    max=get_row_max_by_dataset(combined_data)
    config_columns = max.filter(like='config/').columns.tolist()
    config={}
    for index, row in df_fmax_score.iterrows():
        dataset=row['dataset']        
        dir= row['dir']
        model=row['model']
        percentage=row['percentage']
        
        #print(row['test'])
        #print(row['n_transformations'])
        n_transformations=row['n_transformations']
        folder_=f'../execute_once_experiments/{dataset}/{test}/configs/'
        create_folder(folder_)
        if row['test']=="sep_transf":
            file=f'../{dir}{model}/{dataset}/{model}_{n_transformations}T_{dataset}_P{percentage}/base_config.yaml'
        else:            
            file=f'../{dir}/{dataset}/{model}_{dataset}_P{percentage}/base_config.yaml'
        print(file)
        print(row['score'])

        new_file=f'{folder_}/{model}_{dataset}_P{percentage}.yaml'
        for config_column in config_columns:  
            if 'latent_dim' in config_column:
                config['n_components']=row[config_column]
            elif 'transform_funcs' in config_column:
                values=row[config_column]
                print(f"Tipo de datos: {type(values)}")
                
                if isinstance(values, float):
                    config[config_column.split('/')[1]]=row[config_column]
                else:
                    numeric_values = re.findall(r'\d+', values)
                    transform_funcs = [int(valor) for valor in numeric_values]
                    config[config_column.split('/')[1]] = transform_funcs
            else:
                config[config_column.split('/')[1]]=row[config_column]

                
        new_base_config(file,config,new_file)
            
 
    
    
```

```python
test='all_models_percentage'
df_fmax_score=get_df_fmax_score(['dataset','model', 'percentage'])
combined_data=get_combined_data(df_fmax_score)
create_best_config_exploration(test)
```
