```python
def read_csv(dataset,type_simclr,n_transformations,percentage,dir):
    
    try:
        if(n_transformations>-1):
            folder=f'../{dir}{type_simclr}/{dataset}/'
            folder_percentage=f'{type_simclr}_{n_transformations}T_{dataset}_P{percentage}/data.csv'
        else:
            folder=f'../{dir}/{dataset}/'
            folder_percentage=f'{type_simclr}_{dataset}_P{percentage}/data.csv'
            #print('n_transformations',n_transformations)
            
        
        data = pd.read_csv(f'{folder}{folder_percentage}')
        data.loc[data['score'] < 0,'score'] = 0
    except:
        print(f'{folder}{folder_percentage}')
    
        data = pd.DataFrame({'Unnamed: 0': [], 'score': []})
    return data
    
```

```python
import yaml
import pandas as pd
import pandas as pd
datasets = ['kuhar', 'motionsense', 'uci',  'realworld_thigh']
#datasets = [ 'uci']

models=['simclr_linear','simclr','simclr_full']
percentages=[25, 50, 75, 100, 200]
n_transformations_list=[1,2,3,4]
def get_max_by_each_transformation():
    dir="experiments/simclr_all/by_n_transformations/"
    
    df_maximas = pd.DataFrame() 
    for dataset in datasets:       
            for i, percentage in enumerate(percentages):
                for n_transformations in n_transformations_list:
                    max_length = 0
                    for j, model in enumerate(models):
                        try:
                            data = read_csv(dataset, model, n_transformations, percentage,dir)
                            max_dataset_score = data['score'].max()
                            indice_maximo = data['score'].idxmax()
                            fila_maxima = data.loc[indice_maximo].copy()  # Create a copy to avoid SettingWithCopyWarning
                            fila_maxima['percentage']=percentage
                            fila_maxima['n_transformations']=n_transformations
                            fila_maxima['dir']=str(dir)
                            fila_maxima['dataset']=dataset
                            fila_maxima['model']=model
                            fila_maxima['test']='sep_transf'

                            df_maximas = pd.concat([df_maximas, fila_maxima.to_frame().T], ignore_index=True)
    
                        except:
                            e=1
    return df_maximas
           

df_each_tranf=get_max_by_each_transformation()
    
    
                   
df_each_tranf                
```

```python
def get_max_all_transf():
    test_list=['tem_transf','best_transf','tem_transf']
    #test_list=['tem_transf']
    df_maximas = pd.DataFrame()       
    for dataset in datasets:           
            for i, percentage in enumerate(percentages):
             for model in models:
                for test in test_list:
                    dir=f'experiments/simclr_all/all_transformations/{model}/{test}'
    
                    try:                
                        n_transformations=-1
                        data = read_csv(dataset, model, n_transformations, percentage,dir)
                        max_dataset_score = data['score'].max()
                        indice_maximo = data['score'].idxmax()
                        fila_maxima = data.loc[indice_maximo].copy()  # Create a copy to avoid SettingWithCopyWarning
                        fila_maxima['percentage']=percentage
                        fila_maxima['n_transformations']=n_transformations
                        fila_maxima['dir']=str(dir)
                        fila_maxima['dataset']=dataset
                        fila_maxima['model']=model  
                        fila_maxima['test']=test
                        
                        df_maximas = pd.concat([df_maximas, fila_maxima.to_frame().T], ignore_index=True)
                   
                    except:
                       e=1
    return df_maximas
df_all_tf=get_max_all_transf()
df_all_tf
```

```python
idx_max = df_all_best.groupby(['dataset','model', 'percentage'])['score'].idxmax()
df_fmax_score = df_all_best.loc[idx_max]

combined_data = pd.DataFrame()
for index, row in df_fmax_score.iterrows():
    dataset=row['dataset']
    model=row['model']
    n_transformations=row['n_transformations']
    percentage=row['percentage']
    dir=row['dir']
    data = read_csv(dataset, model, n_transformations, percentage,dir)
    data['dataset'] = dataset
    data['model'] = model
    data['n_transformations'] = n_transformations
    data['percentage'] = percentage
    combined_data = pd.concat([combined_data, data], ignore_index=True)
combined_data.columns
combined_data.to_csv(f'results.csv', index=False)
```
