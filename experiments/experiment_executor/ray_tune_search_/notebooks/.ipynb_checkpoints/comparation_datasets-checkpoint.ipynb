```python
import pandas as pd
import yaml

df = pd.read_csv('results_simclr.csv')
df.loc[df['score'] < 0,'score'] = 0

def get_no_reducer_score(dataset):
    with open(f'../TV_sb_no_reducer/scores/no_reducer_{dataset}.yaml') as f:
        score = yaml.load(f, Loader=yaml.FullLoader)
        score = score['score']
    return score


datasets = ['kuhar', 'motionsense', 'uci', 'wisdm', 'realworld_thigh', 'realworld_waist']
score_no_reducer={}
for dataset in datasets:
   score_no_reducer[dataset] =get_no_reducer_score(dataset)
    
    
```

```python
import seaborn as sns
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go

lista_colores = ["purple", "#e31a1c", "green", "gray", "yellow","blue"]








def box_plot_datasets(df,x):
    dummy_fig = px.box(df.head(0), x=x, y='score', color='dataset', color_discrete_sequence=lista_colores)
    box_colors = [trace.marker.line.color for trace in dummy_fig.data]

    fig = px.box(df, x=x, y='score', color='dataset',
                 title=f'Score vs {x}',
                 labels={'score': 'Score', 'percentage': 'Porcentaje'},
                 color_discrete_sequence=lista_colores,  # Use custom colors
                 width=800, height=500)
    
    cdatasets = df['dataset'].unique()
    
    for dataset, color in zip(datasets, lista_colores):    
        fig.add_shape(
            go.layout.Shape(
                type="line",
                x0=df[x].min(),
                x1=df[x].max(),
                y0=score_no_reducer[dataset],
                y1=score_no_reducer[dataset],
                line=dict(color=color, width=2, dash='dash'),
            )
        )
    
    fig.show()

box_plot_datasets(df,'percentage')
box_plot_datasets(df,'model')
```

```python
import seaborn as sns
import plotly.express as px
import pandas as pd

# Supongamos que tienes un DataFrame llamado combined_data

# Aplica cummax() a la columna 'score' por cada grupo 'dataset'
combined_data['score_cummax'] = combined_data.groupby(['dataset','percentage'])['score'].max()

# Crea el gráfico de línea con la nueva columna 'score_cummax'
fig = px.line(combined_data, x='percentage', y='score_cummax', color='dataset',
              title='Evolución de Score (cummax) por Dataset y Porcentaje',
              labels={'score_cummax': 'Score (cummax)', 'percentage': 'Porcentaje'},
              width=800, height=500)

fig.show()
combined_data['score_cummax']
```

```python
import pandas as pd
import plotly.graph_objects as go

# Assuming combined_data is your DataFrame
# Replace 'your_dataframe.csv' with the actual file path if loading from a CSV file
# combined_data = pd.read_csv('your_dataframe.csv')

def generate_combined_plot(dataset, percentage, model):
    data = combined_data[(combined_data['dataset'] == dataset) & 
                         (combined_data['percentage'] == percentage) & 
                         (combined_data['model'] == model) ]
    
    max_dataset_score = data['score'].cummax()
    trace = go.Scatter(x=list(range(len(max_dataset_score))), y=max_dataset_score, mode='lines', name=f'{percentage}%')
    
    return trace

fig = go.Figure()

# Add each trace to the same Figure object
fig.add_trace(generate_combined_plot('uci', 25, 'simclr'))
fig.add_trace(generate_combined_plot('uci', 50, 'simclr'))
fig.add_trace(generate_combined_plot('uci', 75, 'simclr'))
fig.add_trace(generate_combined_plot('uci', 100, 'simclr'))
fig.add_trace(generate_combined_plot('uci', 200, 'simclr'))

# Set plot labels and title
fig.update_layout(
    xaxis_title='Index',
    yaxis_title='Cumulative Maximum Score',
    title='Combined Plot of Cumulative Maximum Score for different Percentages'
)

# Show the plot
fig.show()
```
